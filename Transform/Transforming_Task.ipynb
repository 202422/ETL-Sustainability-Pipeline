```python
# Librairies import
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import reverse_geocoder as rg
from sklearn.preprocessing import StandardScaler
```

```python
# Dataset Import
df = pd.read_csv("weather_data_cleaned.csv")
df
```

```python
df.info()
```

```python
# Converting column "time" into datetime format and "weather_code" into category
df["time"] = pd.to_datetime(df["time"], format="%Y-%m-%d %H:%M:%S")
df["weather_code"] = df["weather_code"].astype("category")

df.info()
```

## Creating location column

```python
# Extracting locations
locs = df[["latitude", "longitude"]].drop_duplicates().values
locs
```

```python
# Perform reverse geocoding
coordinates = [(lat, lon) for lat, lon in locs]
results = rg.search(coordinates)
results
```

```python
# Create DataFrame with location data
location_data = [
    {
        "latitude": lat,
        "longitude": lon,
        "location": f"{res['name']}"
    }
    for (lat, lon), res in zip(locs, results)
]
location_df = pd.DataFrame(location_data)
location_df
```

```python
# Merge with the original DataFrame (df_all_imputed)
df_with_locations = df.merge(
    location_df,
    on=["latitude", "longitude"],
    how="left"
)
# Converting column "time" into datetime format,"weather_code" and "location" into category
df_with_locations["time"] = pd.to_datetime(df_with_locations["time"], format="%Y-%m-%d %H:%M:%S")
df_with_locations["weather_code"] = df_with_locations["weather_code"].astype("category")
df_with_locations["location"] = df_with_locations["location"].astype("category")

df_with_locations.info()
```

```python
df_with_locations.info()
```

## Time-of-Day Aggregation

```python
def transform(X):
        X_cop = X.copy()
        # Define time-of-day bins
        bins = [0, 6, 12, 18, 24]
        labels = ['Night', 'Morning', 'Afternoon', 'Evening']
        X_cop['time_of_day'] = pd.cut(X_cop['time'].dt.hour, bins=bins, labels=labels, right=False, include_lowest=True)
        X_cop['date'] = X_cop['time'].dt.date
        agg_dict = {
            'temperature_2m': 'mean',
            'relative_humidity_2m': 'mean',
            'precipitation': 'mean',
            'wind_speed_10m': 'mean',
            'weather_code': lambda x: x.mode()[0],
            'location': 'first'  # Preserve location as category
        }
        # Perform aggregation and reset index
        result = X_cop.groupby(['latitude', 'longitude', 'date', 'time_of_day'], observed = True).agg(agg_dict).reset_index()
        # Ensure dtypes are preserved
        result['time_of_day'] = result['time_of_day'].astype('category')
        result['weather_code'] = result['weather_code'].astype('category')
        result['location'] = result['location'].astype('category')
        return result
```

```python
df_TimeOfDay = transform(df_with_locations)
df_TimeOfDay
```

## Creating columns "heat_index" and "is_raining"

```python
# fonction qui calcule une nouvelle colonne "heat_index"
def calculate_heat_index(row):
    T = row['temperature_2m']
    RH = row['relative_humidity_2m']
    if T >= 20:
        return T + 0.33 * RH - 0.7
    return T
```

```python
df_TimeOfDay['heat_index'] = df_TimeOfDay.apply(calculate_heat_index, axis=1)
df_TimeOfDay
```

```python
# création d'une nouvelle colonne "is_raining"
df_TimeOfDay['is_raining'] = (df_TimeOfDay['precipitation'] > 0).astype(int)
df_TimeOfDay
```

## Data Normalization

```python
columns_to_normalize = ['temperature_2m', 'relative_humidity_2m', 'precipitation', 'wind_speed_10m', 'heat_index']
```

```python
def normalize_data(X, columns_to_normalize):
    """
    Normalise les colonnes spécifiées avec StandardScaler (Z-score).
    
    Parameters:
    - X: DataFrame pandas
    - columns_to_normalize: Liste des colonnes à normaliser
    
    Returns:
    - DataFrame avec les colonnes normalisées
    """
    X = X.copy()
    scaler = StandardScaler()
    
    # Vérifier que les colonnes existent
    missing_cols = [col for col in columns_to_normalize if col not in X.columns]
    if missing_cols:
        raise ValueError(f"Colonnes manquantes : {missing_cols}")
    
    print("\nAvant normalisation :")
    print(X[columns_to_normalize].describe())
    
    # Appliquer la normalisation
    X[columns_to_normalize] = scaler.fit_transform(X[columns_to_normalize])
    
    print("\nAprès normalisation :")
    print(X[columns_to_normalize].describe())
    
    return X
```

```python
df_normalized = normalize_data(df_TimeOfDay, columns_to_normalize)
```
